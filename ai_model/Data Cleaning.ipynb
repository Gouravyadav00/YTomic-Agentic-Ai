```python
import pandas as pd
import re
```

```python
df=pd.read_csv("Data.csv")
df.head()
```

```python
def extract_time(text):
    if pd.isna(text):
        return None
    return text[1:6] if text else None

# Create empty lists to store timestamps
start_times = []
stop_times = []

# Iterate through each row
for i in range(len(df)):
    current_time = extract_time(df.iloc[i][0])
    next_time = extract_time(df.iloc[i+1][0]) if i < len(df)-1 else None
    
    start_times.append(current_time)
    stop_times.append(next_time)

# Add the columns to the dataframe
df['start'] = start_times
df['stop'] = stop_times

# Display the result
df[['start', 'stop']].head()
```

```python
#  Remove columns 1, 2, and 3
df = df.drop(columns=['1', '2', '3'])

# Remove first 6 characters (timestamp) from column 0
df['0'] = df['0'].str[7:]

# Display the result
df.head()
```
